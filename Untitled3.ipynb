```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
red_wine_df = pd.read_csv('/content/winequality-red.csv')
white_wine_df = pd.read_csv('/content/winequality-white.csv')
red_wine_df
```

```python
fig, axes = plt.subplots(nrows=4, ncols=6, figsize=(30, 20))

#plots for red wine data,it shows the distribution of each feature and target variable.
#first two rows are for red wine data.
for i in enumerate(red_wine_df.columns[:6]):
    sns.histplot(red_wine_df[i[1]], ax=axes[0,i[0]], kde=True, color='red', bins=20)
for i in enumerate(red_wine_df.columns[6:]):
    sns.histplot(red_wine_df[i[1]], ax=axes[1,i[0]], kde=True, color='red', bins=20)

#plots for white wine data,it shows the distribution of each feature and target variable.
# third and foorth rows for white wine data.
for i in enumerate(white_wine_df.columns[:6]):
    sns.histplot(white_wine_df[i[1]], ax=axes[2,i[0]], kde=True, color='red', bins=20)
for i in enumerate(white_wine_df.columns[6:]):
    sns.histplot(white_wine_df[i[1]], ax=axes[3,i[0]], kde=True, color='red', bins=20)
```
